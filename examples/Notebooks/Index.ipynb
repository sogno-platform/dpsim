# Welcome to DPsim
Here, you can find useful documentation and Jupyter Notebook examples to get started with DPsim.

## Contents

- **Documentation**
  - [DPsim Simulator](https://dpsim.fein-aachen.org/)
  - [VILLASframework](https://villas.fein-aachen.org/doc)
  - [Jupyter](https://jupyter.org/documentation)
  - [Python](https://docs.python.org/3.7/)


- **Examples**

  Below you will find a list of representative DPsim examples. Feel free to explore all the examples in the [Notebooks folder](.)
  - [Quickstart Guide](./Quickstart%20Guide.ipynb) Gives a short introduction to the simulator in a simplified linear circuit
  - [Infinite bus system with Fault](./Circuits/DP_SP_SynGenTrStab_SMIB_Fault.ipynb) Shows how to implement the Kundur's SMIB system with a balanced fault from scratch and simulates it using Dynamic Phasors
  - [Three bus system in steady state](./Circuits/DP_SP_SynGenTrStab_3Bus_SteadyState.ipynb) Shows how to implement a power system with multiple buses and generators from scratch and simulates it in dynamic steady state
  - [Dynamic Simulation from CIM model](./Grids/DP_CIGRE_MV_withDG_withLoadStep.ipynb) Demonstrates how to implement a dynamic simulation by importing a grid model defined by the Common Information Model (CIM) specification
  - [Power flow from MATPOWER model](./matpower-case9.ipynb) Demonstrates how to implement a power flow by importing a grid model defined in [MATPOWER](https://matpower.org/)
